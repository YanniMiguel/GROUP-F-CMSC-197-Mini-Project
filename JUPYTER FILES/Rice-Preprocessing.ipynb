```python
size=len(df['date'])
year_arr=np.zeros((size,1))
month_arr=np.zeros((size,1))
i=0
for date in df['date']:
  darr=date.split('-')
  year_arr[i]=darr[0]
  month_arr[i]=darr[1]
  #print(darr[1])
  i+=1

df['year']=year_arr.astype(int)
df['month']=month_arr.astype(int)
df=df.drop(labels='date',axis=1) 

df
```

```python
def uniq(df, column):
    uniq = df[column].unique()
    return uniq
```

```python
def dicts(uniq):

    dict ={}

    for i in range (len(uniq)):
        dict[uniq[i]] = i+1
    return dict
```

```python
def cat_to_num(df, column, dicts):
    size=len(df[column])
    catg=df[column].unique()

    catgarr=list(dicts.keys())
    carr=dicts
    
    for i in range (len(catg)):
      a=np.zeros((size,1))
      exec('catgarr[i] = a')

    for j in range (size):
      cval=df[column][j]
      ind=carr.get(cval)
      catgarr[ind-1][j]=1

    r=0
    for c in catg:
      df[c]=catgarr[r].astype(int)
      r+=1

    
```

```python
prov_uniq = uniq(df2, 'province')
dict_prov = dicts(prov_uniq)
cat_to_num(df2,'province', dict_prov)
df2=df2.drop(labels='province',axis=1) 
df2
```

```python
df2['month'] = df2['month'].apply(str)

month_uniq = uniq(df2, 'month')
month_uniq
```

```python
dict_month = dicts(month_uniq)
cat_to_num(df2,'month', dict_month)
df2=df2.drop(labels='month',axis=1) 
df2
```

```python
dist_uniq = uniq(df2, 'district')
dict_dist = dicts(dist_uniq)
cat_to_num(df2,'district', dict_dist)
df2=df2.drop(labels='district',axis=1) 
df2
```

```python
mrkt_uniq = uniq(df2, 'market')
dict_mrkt = dicts(mrkt_uniq)
cat_to_num(df2,'market', dict_mrkt)
df2=df2.drop(labels='market',axis=1) 
df2
```

```python
pf_uniq = uniq(df2, 'priceflag')
dict_pf = dicts(pf_uniq)
cat_to_num(df2,'priceflag', dict_pf)
df2=df2.drop(labels='priceflag',axis=1) 
df2
```
